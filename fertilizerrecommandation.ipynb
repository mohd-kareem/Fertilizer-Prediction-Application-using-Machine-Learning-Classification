```python
import pandas as pd
```

```python
df = pd.read_csv("Fertilizer Recommendation.csv")
df.head()
```

```python
#converting text values into integers for the ml algorithm
st = df["Soil Type"].unique()
#soil type unique values unique function
#unique function gives the unique values 
print(f"Soil Type {st}")

ct = df['Crop Type'].unique()
print(f"Crop Type {ct}")

fn = df['Fertilizer Name'].unique()
print(f"Fertilizer names {fn}")
```

```python
#changing the string to numeric
#the values are assigned alphabetically with the first character
#for that we used label encoder
#categorical data is assigned numeric values.

#if we have too many values , just drop the column
#**IMPORTANT**
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['Soil Type']=le.fit_transform(df['Soil Type'])
df['Crop Type'] = le.fit_transform(df['Crop Type'])
df['Fertilizer Name'] = le.fit_transform(df['Fertilizer Name'])
```

```python

st = df["Soil Type"].unique()
#soil type unique values unique function
print(f"Soil Type {st}")

ct = df['Crop Type'].unique()
print(f"Crop Type {ct}")

fn = df['Fertilizer Name'].unique()
print(f"Fertilizer names {fn}")
```

```python
x = df.drop(columns='Fertilizer Name')
#prints all the columns except fn
x
```

```python
y = df['Fertilizer Name']
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
print(x_test.head())
print(y_test.head())
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import mean_absolute_error


kn = KNeighborsClassifier(n_neighbors=5)
s = SVC()
rf = RandomForestClassifier()
dt = DecisionTreeClassifier()

rf.fit(x_train,y_train)
dt.fit(x_train,y_train)
s.fit(x_train,y_train)
kn.fit(x_train,y_train)

y_pred2 = rf.predict(x_test)
y_pred1 = dt.predict(x_test) #for the user input 
#it prints the output 
y_pred3 = s.predict(x_test)
y_pred4 = kn.predict(x_test)
#for training data set , it gives a lot ouput cuz training data is large

print(f"Decision tree {y_pred1} accuracy is {mean_absolute_error(y_test,y_pred1)}")
print(f"Random Forest {y_pred2} accuracy is {mean_absolute_error(y_test,y_pred2)}")
print(f"Support Vector classifier {y_pred3} accuracy is {mean_absolute_error(y_test,y_pred3)}")
print(f"Kneighbour classifier {y_pred4}accuracy is {mean_absolute_error(y_test,y_pred4)}")
```
